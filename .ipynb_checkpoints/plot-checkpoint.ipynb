```python
all_files = ["all_lower", "all_med", "all_high", "high_to_med"]
name_mapping = {
    "all_lower": "Low dev area",
    "all_med": "Med dev area",
    "all_high": "High dev area",
    "high_to_med": "High to Med dev area",
}
```

```python
max_res = pd.DataFrame()
for f in all_files:
    close_df = pd.read_csv("National_Park "+ f + "_closed-table.csv", skiprows=6)
    open_df = pd.read_csv("National_Park "+ f + "_open-table.csv", skiprows=6)
    mixed_df = pd.read_csv("National_Park "+ f + "_mixed-table.csv", skiprows=6)
    
    all_df = [close_df, open_df, mixed_df]
    
    for df in all_df:
        max_data = find_max(df)
        max_data['class'] = name_mapping[f]
        max_res = max_res.append(max_data)
```

```python
max_res
```

```python
g = sns.catplot(x="strategy", y="revenue", col="class",
                data=max_res, kind="bar",
                height=8, aspect=.4)
g.set_ylabels("Total Profit")
g.savefig("profit.png")
```

```python
g = sns.catplot(x="strategy", y="cumulative-come-tourists / (cumulative-considered-tourists + 1)", col="class",
                data=max_res, kind="bar",
                height=8, aspect=.4, orient="v")
g.set_ylabels("Tourists Flow")
g.savefig("flow.png")
```
